```python
# Create a class called "Calculator" 

class calculator :
    
    dic_oper={}

    def add(self,a,b):
        return a+b

    def sub(self,a,b):
        return a-b

    def mul(self,a,b):
        return a*b

    def div(self,a,b):
        return a/b

    def init(self):
        calculator.dic_oper={
                '+':self.add,
                '-':self.sub,
                '*':self.mul,
                '/':self.div
                  }
        
    def add_operation(self,oper_name,func):
        if not calculator.dic_oper:
            self.init()

        if oper_name not in calculator.dic_oper:
            calculator.dic_oper[oper_name]=func
            return f'operator added'
        else :
            return f'operator exists'

    def calculate(self,first,oper,second=0):
        liste_op=['+','/','*','-']

        if not isinstance(first,(int,float,complex)) or not isinstance(second,(int,float,complex))  :
            return f"one of the values are not numbers !"
            
        if oper in calculator.dic_oper and oper in liste_op:
            return calculator.dic_oper[oper](first,second)
        elif oper in calculator.dic_oper and oper not in liste_op:
            return calculator.dic_oper[oper](first)
        else :
            return f'the operator symbol is not valid !'
```

```python
# create seperate functions (exponentiation, square root, logarithm) and use the "add_operation" to add them

def expo(a):
    return math.exp(a)

def sqrt(a):
    return math.sqrt(a)   

def log(a):
    return math.log(a) 
```

```python
# to check valua type

def type_num(a):
    
    if a.isdigit():
        return int(a) 
    else :
        return float(a)
```

```python
# use the "add_operation" method to add new functions to the calculator's dictionary

calculator().add_operation('expo',expo)
calculator().add_operation('sqrt',sqrt)
calculator().add_operation('log',log)
```
